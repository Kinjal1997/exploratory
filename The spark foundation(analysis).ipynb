```python
data6 = ['Sales','Quantity','Discount','Profit']
data7 = [2.297201e+06,3.787300e+04,1.561090e+03,2.863970e+05]
plt.bar(data6,data7,color='red')
plt.xlabel('total of sales,quantity,discount,profit')
plt.ylabel('sum of all')
plt.title("Histogram of samplesuperstore")
plt.show()
```

```python
data6 = ['Sales','Quantity','Discount','Profit']
data7 = [2.297201e+06,3.787300e+04,1.561090e+03,2.863970e+05]
explode = (0.4,0.4,0.4,0.4)
fig1,ax1=plt.subplots()
ax1.pie(data7,explode=explode,labels=data6,autopct='%1.1f%%',shadow=True,startangle=90)
ax1.axis('equal')
plt.show()
```
